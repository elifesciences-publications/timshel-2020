```python
import numpy as np # needed for formatting data for this tutorial
import pandas as pd # needed for formatting data for this tutorial
import cellex
```

```python
path_data = "../../tmp-data/expression/gtex_v8_filtered_normLog_all.csv.gz"
path_metadata = "../../tmp-data/expression/gtex_v8_filtered_normLog_all.metadata.csv"
path_esmu_out = "../../out/es/"
prefix_data = "gtex_v8_filtered_normLog_all"
```

```python
filename_out = path_esmu_out + prefix_data + ".mu.csv.gz"
```

```python
filename_out
```

```python
data = pd.read_csv(path_data, index_col=0, na_filter=False, low_memory=False)
```

```python
data.head()
```

```python
metadata = pd.read_csv(path_metadata,  index_col=0, na_filter=False, low_memory=False)
```

```python
metadata.head()
```

```python
eso = cellex.ESObject(data=data, annotation=metadata, normalize=False, verbose=True)
eso.compute(verbose=True)
```

```python
eso.results["esmu"].to_csv(filename_out)
```
