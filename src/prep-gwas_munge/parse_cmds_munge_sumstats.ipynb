```python
import os
import sys
import time

import re

import pdb

import numpy as np
import pandas as pd

import glob
```

```python
file_in = "/projects/timshel/sc-genetics/sc-genetics/src/prep-gwas_munge/README-cmds_munge_sumstats.txt"
```

```python
### Parse file
dict_gwas_txt = {}
with open(file_in, "r") as fh_in:
    for count, line in enumerate(fh_in):
        # print(count, line)
        line = line.strip()
        if line.startswith("#"): 
            continue
        if line == "": # this is to avoid parsing problems of the 'looped' munged GWAS
            # ^ if we get a blank line, we are sure we are done parsing the current GWAS
            gwas_id = "irregularity in file"
            dict_gwas_txt[gwas_id] = {} # init dict of dict
        if line.startswith("GWAS="): 
            gwas_id = re.search(r"GWAS=(.*)$", line).group(1)
            dict_gwas_txt[gwas_id] = {} # init dict of dict
        if line.startswith("--sumstats"): 
            tmp_filepath = re.search(r"--sumstats (.*) \\$", line).group(1)
            dict_gwas_txt[gwas_id]["file_input"] = os.path.basename(tmp_filepath)
            dict_gwas_txt[gwas_id]["file_dir"] = os.path.basename(os.path.dirname(tmp_filepath))
        if line.startswith("--N "): # trailing space in "--N " needed to distinguish from "--N-col"
            dict_gwas_txt[gwas_id]["N"] = re.search(r"(\d+)", line).group(1)
            dict_gwas_txt[gwas_id]["N arg parsed"] = True
```

```python

### Create data frame from dict of dicts
### REF: https://stackoverflow.com/questions/33157522/create-pandas-dataframe-from-dictionary-of-dictionaries
df = pd.DataFrame.from_dict(dict_gwas_txt, orient='index')
df.index.name = "gwas_id"
print(df.shape)
df.head()
```
