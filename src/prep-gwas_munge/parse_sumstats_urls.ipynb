```python
import os
import sys
import time

import re

import pdb

import numpy as np
import pandas as pd


import glob
from collections import defaultdict
from pprint import pprint
```

```python
dir_in = "/projects/timshel/sc-genetics/sc-genetics/data/gwas_sumstats_raw/"

pattrn_url = re.compile("URL|www|http|ftp", re.IGNORECASE)
```

```python
dict_urls = defaultdict(list)
for filename in glob.iglob("{}/**/README*.txt".format(dir_in), recursive=True):
    gwas_id = "--".join(filename.split("/")[-2:]) # os.path.basename(os.path.filename), os.path.basename(filename)
    with open(filename, "r") as fh_in:
        for line in fh_in:
            line = line.strip()
            m = re.search(pattrn_url, line)
            if m:
                dict_urls[gwas_id].append(line)
```
